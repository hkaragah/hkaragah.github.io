# SMF BEAM STRENGTH CHECK

## Required Strengths

According to ASCE 7, the following required strength shall be determined by 2nd-order analysis including the effects of $P-\delta$ and $P-\Delta$ with _reduced stiffness_ as required by _Direct Analysis Method_.

* Required strength at the face of column

$$ M_u = (1.2+0.2S_{SD})D+\rho Q_E + 0.5L + 0.15S $$
$$ V_u = (1.2+0.2S_{SD})D+\rho Q_E + 0.5L + 0.15S $$

* Required strength at the centerline of RSB

$$ M_u = (1.2+0.2S_{SD})D+\rho Q_E + 0.5L + 0.15S $$

## RBS Dimensions

According to ASCE 358-22, §5.7, step 1:

* $0.5b_{bf} \le a \le 0.75b_{bf}$ (Eq. 5.7-1)
* $0.65d \le b \le 0.85d$ (Eq. 5.7-2)
* $0.1b_{bf} \le c \le 0.25b_{bf}$ (Eq. 5.7-3)
